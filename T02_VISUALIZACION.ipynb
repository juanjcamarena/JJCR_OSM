```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
```

$$ f(x) = x^2 - 4x + 2$$

$$ 2x -4 = 0$$

$$ 2x = 4 $$

$$ x = 2 $$

```python
x = np.linspace(-5, 5, 100)
y = (x)**2 - 4*x + 2
plt.plot (x, y)
plt.grid()
plt.plot(2, -2, "*r")
```

$$ f(x) = sin(3x) $$

$$ cos(3x) * 3 $$

```python
x = np.linspace(-5 , 5 , 1000)
y = np.sin(3*x)
plt.plot(x, y)
plt.grid()
```

$$ f(x) = e^{2x} -3 $$

```python
x = np.linspace(5, -5, 100)
y = np.exp(2*x) -3
plt.plot(x, y)
plt.grid()
```

$$ f(x) = \sqrt{x^2 -4} $$

```python
x = np.linspace(-5, 5, 1000)
y = np.sqrt(np.maximum((x)**2 - 4, 0))
plt.plot(x, y)
plt.grid()
```

$$ f(x) = cos(x + \frac\pi 4)$$

```python
x = np.linspace(-2 *np.pi, 2*np.pi, 1000)
y = np.cos(x + np.pi /4)
plt.plot(x, y)
plt.grid()
```

$$ y1 = 2x + 1$$

$$ y2 = -x + 4 $$

$$ y3 = x^2$$

```python
x = np.linspace(-3, 6, 100)
y1 = 2*x + 1
y2 = -x + 4
y3 = x**2
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.grid()
```

$$ y1 = sin(x)$$

$$ y2 = cos(x) $$

```python
x = np.linspace(-5, 5, 1000)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1)
plt.plot(x, y2)
plt.grid()
```

```python
def senos(x1, x2, n, a, b, k):
    x = np.linspace(x1, x2, n)
    alphas = (a, b, k)
    for alpha in alphas:
        y = np.sin(alpha * x)
        plt.plot(x, y, label=f' = {alpha:.2f}')
```
